```python
## import the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

### load the data
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
dataset.data
```

```python
dataset.feature_names
```

```python
dataset.target
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df
```

```python
df['Blood Suger']=dataset.target
```

```python
df
```

```python
### let's split the data into train and test
x=df.drop(columns='Blood Suger')
```

```python
x.head()
```

```python
## get y 

y=df.iloc[:,-1]
```

```python
y
```

```python
## check the number or row and columns in my x and y
x.shape,y.shape
```

```python
## importing the module to split the data into train and test
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
### train the model

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error , r2_score
```

```python
linear_model=LinearRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
### check the slop of alow x
linear_model.coef_
```

```python
### check the intercept
linear_model.intercept_
```

```python
### check the error

y_train_pred=linear_model.predict(x_train)
```

```python
y_train
```

```python
y_train_pred
```

```python
## erorr 
mean_squared_error(y_train,y_train_pred)
```

```python
### performance
r2_score(y_train,y_train_pred)
```

```python
### let's check the model performance with my test data

y_test_pred=linear_model.predict(x_test)
```
